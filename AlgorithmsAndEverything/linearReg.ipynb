```python
# Sütun isimlerini manuel veriyoruz , çünkü ilk satır data ve bu veri setinde columns eklenmemiş.Sklearn
# hazır kütüphanesinden de indiremiyoruz.
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 
           'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

df = pd.read_csv('housing.csv', header=None, names=columns, sep='\s+')
df
```

```python
def aykiri_deger_raporu(df):
    rapor = []
    
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        alt_sinir = Q1 - 1.5 * IQR
        ust_sinir = Q3 + 1.5 * IQR
        
        aykirilar = df[(df[col] < alt_sinir) | (df[col] > ust_sinir)][col]
        
        rapor.append({
            'Sütun'        : col,
            'Q1'           : round(Q1, 2),
            'Q3'           : round(Q3, 2),
            'IQR'          : round(IQR, 2),
            'Alt Sınır'    : round(alt_sinir, 2),
            'Üst Sınır'    : round(ust_sinir, 2),
            'Aykırı Sayısı': len(aykirilar),
            'Aykırı %'     : round(len(aykirilar) / len(df) * 100, 2)
        })
    
    return pd.DataFrame(rapor)
aykiri_deger_raporu(df)
```

```python
# chas binary değişken olduğu için listede yok
cols_to_cap = ['crim', 'zn', 'rm', 'dis', 'ptratio', 'b', 'lstat', 'medv']

for col in cols_to_cap:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    alt = Q1 - 1.5 * IQR
    ust = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=alt, upper=ust)

# Kontrol et
rapor_sonra = aykiri_deger_raporu(df)
print(rapor_sonra[['Sütun', 'Aykırı Sayısı', 'Aykırı %']].to_string(index=False))
```

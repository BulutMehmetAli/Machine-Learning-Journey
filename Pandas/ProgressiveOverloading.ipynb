```python
def order(data):
    rank = np.full_like(data , '$', dtype='U3')
    rank[(data >= 10) & (data<30)] = '$$'
    rank[(data >= 30)] = '$$$'
    return rank
```

```python
# Tüm sütunlar için
z_scores = df.select_dtypes(include='number').apply(st.zscore)
print(z_scores)
```

```python
# any(axis=1)Satırda en az 1 True varsa → True
# any(axis=0)Sütunda en az 1 True varsa → True
# all(axis=1)Satırın tamamı True ise → True
# .any(axis=1) Ne Yapar?
# "Her satırda en az bir True var mı?" diye sorar:
# 0  → False False False → False
# 1  → False False False → False
# 2  → True  False False → True  ✅ outlier!
# 3  → False True  False → True  ✅ outlier!
# axis=1 ise satır boyunca bak demek, axis=0 olsaydı sütun boyunca bakardı.
```

```python
z_score = df.select_dtypes(include='number').apply(st.zscore)
z_score
```

```python
z_score = df.select_dtypes(include='number').apply(st.zscore)

# Herhangi bir sütunda z_score > 3 olan satırları getir
outlier_mask = (abs(z_score) > 3).any(axis=1)
outliers = df[outlier_mask]

print(f"Toplam outlier sayısı: {len(outliers)}")
outliers
```

```python
z_score = df.select_dtypes(include='number').apply(st.zscore)

# Outlier olmayan satırlar
outlier_mask = (abs(z_score) > 3).any(axis=1)
temiz_df = df[~outlier_mask]  # ~ işareti "değil" anlamına gelir

print(f"Toplam satır: {len(df)}")
print(f"Outlier sayısı: {outlier_mask.sum()}")
print(f"Temiz veri sayısı: {len(temiz_df)}")
temiz_df
```

```python
z_score = df.select_dtypes(include='number').apply(st.zscore)
outlier_mask = (abs(z_score) > 3).any(axis=1)
temiz_df = df[~outlier_mask]

# Her sütunun outlier olmayan aralığı
for sütun in temiz_df.select_dtypes(include='number').columns:
    print(f"{sütun} → Min: {temiz_df[sütun].min():.2f} | Max: {temiz_df[sütun].max():.2f}")
```

```python
from scipy import stats

z_scores = stats.zscore(df['price_per_person'])

alt_sınır = df['price_per_person'][abs(z_scores) <= 3].min()
üst_sınır = df['price_per_person'][abs(z_scores) <= 3].max()

print(f"Alt sınır: {alt_sınır:.2f}")
print(f"Üst sınır: {üst_sınır:.2f}")
```
